```python
import pandas as pd
from glob import glob
import matplotlib.pyplot as plt
import seaborn as sb
import requests
from bs4 import BeautifulSoup
import json
import pymongo
```

```python
dataframes = []

for i in range(1, 13):
    nombre_csv = f'/workspaces/myFirstRep-ID/bienes_json/ReportePCBienes2022{i:02d}.csv'
    nombre_json = f'Bienes_ReportePC_2022{i:02d}.json'

    try:
        df = pd.read_csv(nombre_csv, encoding='latin-1', delimiter=';')
        diccionario_resultante = df.to_dict(orient='records')

        with open(nombre_json, 'w') as archivo_json:
            json.dump(diccionario_resultante, archivo_json)

        print(f"Conversión completa. Archivo JSON guardado como '{nombre_json}'.")
    except FileNotFoundError:
        print(f"Archivo {nombre_csv} no encontrado.")
```
