```python
directorio = "D:/raul/ETL completo - practica - S12/archivos csv/"

patron = "ReportePCServicios*.csv"

archivos = glob(directorio + patron)
```

```python
dataframes = {}
for archivo in archivos:
    nombre_variable = "r_" + archivo[-10:-4]
    dataframes[nombre_variable] = pd.read_csv(archivo, encoding='latin-1', delimiter=';', dtype={'RUC_PROVEEDOR': 'str', 'RUC_ENTIDAD': 'str'})
```

```python
dataframes["r_202201"].head(3)
```

```python
for nombre_df, df in dataframes.items():
    dataframes[nombre_df] = df.drop(['FECHA_PROCESO', 'ACUERDO_MARCO'], axis=1)
```

```python
for nombre_df, df in dataframes.items():
    plt.figure(figsize=(35, 8))
    sb.heatmap(df.isnull(), cbar=False, cmap='inferno')
    plt.xticks(rotation=0)
    plt.title(f'Mapa de calor para valores nulos en {nombre_df}')
    plt.show()
```

```python
total_gasto_mensual = {}

for nombre_df, df in dataframes.items():
    # Extraer el año y mes del nombre del DataFrame
    year_month = nombre_df.split("_")[1]
    year = year_month[:4]
    month = year_month[4:]

    # Sumar la columna 'TOTAL' de cada DataFrame
    total_gasto = df['TOTAL'].sum()

    # Almacenar los resultados en el diccionario
    total_gasto_mensual[f"{month}/{year}"] = total_gasto

# Imprimir los resultados
for fecha, total_gasto in total_gasto_mensual.items():
    print(f"Gasto total en {fecha}: {total_gasto}\n")
```

```python
proveedores_mas_utilizados = {}

for nombre_df, df in dataframes.items():
    # Extraer el año y mes del nombre del DataFrame
    year_month = nombre_df.split("_")[1]
    year = year_month[:4]
    month = year_month[4:]

    # Calcular la frecuencia de los proveedores
    frecuencia_proveedores = df['PROVEEDOR'].value_counts()

    # Almacenar los resultados en el diccionario
    proveedores_mas_utilizados[f"{month}/{year}"] = frecuencia_proveedores

# Imprimir los resultados
for fecha, frecuencia_proveedores in proveedores_mas_utilizados.items():
    print(f"Proveedores más utilizados en {fecha}:\n{frecuencia_proveedores}\n")
```

```python
actividad_mensual = {}

for nombre_df, df in dataframes.items():
    # Extraer el año y mes del nombre del DataFrame
    year_month = nombre_df.split("_")[1]
    year = year_month[:4]
    month = year_month[4:]

    # Contar las filas de cada DataFrame
    actividad_mensual[f"{month}/{year}"] = len(df)

# Imprimir los resultados
for nombre_df, actividad in actividad_mensual.items():
    print(f"Actividad en {nombre_df}: {actividad} compras\n")
```

```python
# Inicializar listas para almacenar datos
meses = []
gastos_totales = []

# Iterar a través del diccionario de dataframes
for nombre_df, df in dataframes.items():
    # Extraer el año y mes del nombre del DataFrame
    year_month = nombre_df.split("_")[1]
    year = year_month[:4]
    month = year_month[4:]

    # Guardar el mes y año
    meses.append(f"{month}/{year}")

    # Sumar la columna 'TOTAL' de cada DataFrame
    total_gasto = df['TOTAL'].sum()
    gastos_totales.append(total_gasto)

# Crear el gráfico de línea
plt.figure(figsize=(10, 6))
plt.plot(meses, gastos_totales, marker='o', linestyle='-', color='b')
plt.title('Tendencia del Gasto Total a lo largo de los Meses')
plt.xlabel('Meses')
plt.ylabel('Gasto Total')
plt.xticks(rotation=45, ha='right')  # Rotar etiquetas del eje x para mayor claridad
plt.grid(True)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# Crear un DataFrame con la frecuencia de uso de proveedores
proveedores_frecuencia = pd.DataFrame(columns=['Proveedor', 'Frecuencia'])

for nombre_df, df in dataframes.items():
    frecuencia_proveedores = df['PROVEEDOR'].value_counts()
    proveedores_frecuencia = pd.concat([proveedores_frecuencia, pd.DataFrame({'Proveedor': frecuencia_proveedores.index, 'Frecuencia': frecuencia_proveedores.values})])

# Agrupar por proveedor y sumar la frecuencia total
proveedores_frecuencia_total = proveedores_frecuencia.groupby('Proveedor')['Frecuencia'].sum().sort_values(ascending=False)

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
sb.barplot(x=proveedores_frecuencia_total.values, y=proveedores_frecuencia_total.index, palette='viridis')
plt.title('Frecuencia de Uso de Proveedores')
plt.xlabel('Frecuencia')
plt.ylabel('Proveedor')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
actividad_mensual = pd.DataFrame(columns=['Mes/Año', 'Compras'])

for nombre_df, df in dataframes.items():
    # Extraer el año y mes del nombre del DataFrame
    year_month = nombre_df.split("_")[1]
    year = year_month[:4]
    month = year_month[4:]

    # Contar las filas de cada DataFrame
    actividad_mensual = pd.concat([actividad_mensual, pd.DataFrame({'Mes/Año': [f"{month}/{year}"], 'Compras': [len(df)]})])

# Ordenar el DataFrame por Mes/Año
actividad_mensual = actividad_mensual.sort_values('Mes/Año')

# Crear el gráfico de barras o de línea
plt.figure(figsize=(10, 6))
plt.plot(actividad_mensual['Mes/Año'], actividad_mensual['Compras'], marker='o', linestyle='-')
plt.title('Actividad Mensual')
plt.xlabel('Mes/Año')
plt.ylabel('Número de Compras')

# Puedes elegir entre gráfico de barras o de línea descomentando una de las siguientes líneas:
# plt.bar(actividad_mensual['Mes/Año'], actividad_mensual['Compras'], color='skyblue')

# Rotar las etiquetas del eje x para mayor legibilidad
plt.xticks(rotation=45, ha='right')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
